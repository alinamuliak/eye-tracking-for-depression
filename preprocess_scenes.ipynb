```python
import pandas as pd
import os
```

```python
for file in os.listdir("data/innovaid_hackathon_anima/input"):
    print(file)
    df = pd.read_csv(f"data/innovaid_hackathon_anima/input/{file}").drop("Unnamed: 0", axis=1)
    df["IMAGE_TYPE"] = df["IMAGE_TYPE"].fillna('NaN')

    res = []
    for idx in df["SCENE_INDEX"].unique():
        temp = {}
        dff = df[df["SCENE_INDEX"] == idx]

        values = list(dff["IMAGE_TYPE"].unique())

        if "positive" in values:
            pos = round(dff["IMAGE_TYPE"].value_counts()["positive"] / dff.shape[0], 2)
        else:
            pos = 0.0

        if "negative" in values:
            neg = round(dff["IMAGE_TYPE"].value_counts()["negative"] / dff.shape[0], 2)
        else:
            neg = 0
        
        if "neutral" in values:
            neu = round(dff["IMAGE_TYPE"].value_counts()["neutral"] / dff.shape[0], 2)
        else:
            neu = 0
        
        if "NaN" in values:
            nan = round(dff["IMAGE_TYPE"].value_counts()["NaN"] / dff.shape[0], 2)
        else:
            nan = 0

        res.append([idx, pos, neg, neu, nan])
    
    df_clean = pd.DataFrame(res, columns = ["scene_index", "pos", "neg", "neu", "nan"])

    df_clean.to_csv(f"preprocessed/{file}", index=False)
```
