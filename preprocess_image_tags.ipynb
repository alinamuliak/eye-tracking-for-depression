```python
import pandas as pd
import os
import spacy
```

```python
directory = "data/innovaid_hackathon_anima/input"
all_tags = set()
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        try:
            tags = pd.read_csv(f, usecols=["IMAGE_TAGS"])
            all_tags |= set(tags["IMAGE_TAGS"].unique())
        except ValueError:
            df = pd.read_csv(f)
            print(filename)
            print(df.columns)
```

```python
all_tags = {x for x in all_tags if x==x}
```

```python
nlp = spacy.load('en_core_web_sm')
to_add = {}
for tag in all_tags:
    doc = nlp(tag)
    tokens_str = sorted([token.text for token in doc if not token.is_punct])
    tokens_ids = sorted([token.orth for token in doc if not token.is_punct])
    to_add[tag] = (tokens_str, tokens_ids)
    print(tokens_str)
```

```python
to_add
```

```python
def map_image_tag(tag, tag_mapping):
    if tag in tag_mapping:
        token, token_id = tag_mapping[tag]
        return pd.Series({'TOKENS': token, 'TOKEN_IDS': token_id})
    else:
        return pd.Series({'TOKENS': None, 'TOKEN_IDS': None})

for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        try:
            df = pd.read_csv(f)
#             df["IMAGE_TAGS"]
            df[['TOKENS', 'TOKEN_IDS']] = df['IMAGE_TAGS'].apply(lambda tag: map_image_tag(tag, to_add))
            print(df)
            break
        except ValueError:
            df = pd.read_csv(f)
            print(filename)
            print(df.columns)
```
