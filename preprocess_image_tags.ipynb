```python
import pandas as pd
import os
import spacy
```

```python
directory = "data/innovaid_hackathon_anima/input"
all_tags = set()
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        try:
            tags = pd.read_csv(f, usecols=["IMAGE", "IMAGE_TYPE", "IMAGE_TAGS"])
            tags_list = tags[["IMAGE", "IMAGE_TYPE", "IMAGE_TAGS"]].drop_duplicates().dropna().values.tolist()
            for l in tags_list:
                l.append(filename)
#             print(tags_list)
            unique_tuples = map(tuple, tags_list)
            all_tags.update(unique_tuples)
#             result_dict = df.set_index("IMAGE")["IMAGE_TAGS"].to_dict()
#             print(unique_tuples)
#             all_tags |= set(tags["IMAGE_TAGS"].unique())
        except (ValueError, KeyError):
            df = pd.read_csv(f)
            print(filename)
            print(df.columns)
```

```python
all_tags_lst = list(all_tags)
    
# Sort the list of tuples by the first element
sorted_tuples = sorted(all_tags_lst, key=lambda x: x[0])
# sorted_tuples

def create_dict_from_tuples(tuple_list):
    result_dict = {}

    for tpl in tuple_list:
        key = tpl[:3]  # Extract the first three elements as the key
        value = tpl[3]  # Extract the fourth element as the value

        # Check if key is already in the dictionary
        if key in result_dict:
            result_dict[key].append(value)
        else:
            # If key is not in the dictionary, create a new entry
            result_dict[key] = [value]

    return result_dict

res_dict = create_dict_from_tuples(sorted_tuples)
for el in res_dict:
    print(el, len(res_dict[el]))
```

```python
def transform_dict(original_dict):
    transformed_list = []

    for key, value_list in original_dict.items():
        first_element = key[0]
        second_and_third_elements = (key[1], key[2])
        value_list_length = len(value_list)

        transformed_list.append((first_element, second_and_third_elements, value_list_length))

    return transformed_list

image_ids_dct = transform_dict(res_dict)
image_ids_dct
```

```python
from itertools import groupby

def group_and_filter(input_list):
    # Sort the list by the first element
    sorted_list = sorted(input_list, key=lambda x: x[0])

    # Group the sorted list by the first element
    grouped_by_first_element = groupby(sorted_list, key=lambda x: x[0])

    result_list = []

    for _, group in grouped_by_first_element:
        # For each group, find the entry with the greatest last element
        max_entry = max(group, key=lambda x: x[-1])
        result_list.append(max_entry)

    return result_list

correct_ids = group_and_filter(image_ids_dct)
correct_ids_small = {el[0]: el[1] for el in correct_ids}
correct_pairs = [(el[0], el[1][0], el[1][1]) for el in correct_ids]
correct_pairs
```

```python
import shutil

def copy_directory(source_path, destination_path):
    try:
        # Copy the entire directory from source to destination
        shutil.copytree(source_path, destination_path)
        print(f"Directory '{source_path}' copied to '{destination_path}' successfully.")
    except Exception as e:
        print(f"Error copying directory: {e}")

# Example usage:
source_directory = "data/innovaid_hackathon_anima/input"
destination_directory = "data/innovaid_hackathon_anima/fixed_input"

copy_directory(source_directory, destination_directory)
```

```python
def update_csv(input_file, img_id, new_image_type, new_image_tags):
    # Read the CSV file into a DataFrame
    df = pd.read_csv(input_file)

    # Identify rows where the "IMAGE" column has a specific value
    mask = df['IMAGE'] == img_id

    # Update the "IMAGE_TYPE" and "IMAGE_TAGS" columns for the selected rows
    df.loc[mask, 'IMAGE_TYPE'] = new_image_type
    df.loc[mask, 'IMAGE_TAGS'] = new_image_tags

    # Save the modified DataFrame to a new CSV file
    df.to_csv(input_file, index=False)
```

```python
for i, el in enumerate(res_dict):
    print(f"{i} / {len(res_dict)}")
    if el not in correct_pairs:
        filenames = res_dict[el]
        for filename in filenames:
            f = os.path.join(destination_directory, filename)
            img_id = el[0]
            img_type, img_tags = correct_ids_small[img_id]
            update_csv(f, img_id, img_type, img_tags)
```
